```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
```

```python
url = 'http://jtbc.joins.com/search/tv?term=%EB%8B%A4%ED%81%90'
html = urlopen(url)
soup = BeautifulSoup(html, "lxml")
```

# 수집하고자 하는 데이터를 class명을 이용해서 구체적으로 찾아보기

```python
print(soup.find_all('div', 'wrap_quick_prg'))
```

```python
print(soup.find_all('div', 'box_prg'))
```

```python
title_elements=soup.select('div.box_prg h3.prg_ttl a.txt_ttl')
title_elements
```

```python
titles=[]
for e in title_elements:
    titles.append(e.text)
titles
```

```python
import pandas as pd

data = pd.DataFrame(
    {'title': titles},
    columns={'title'})
data
```

# 다큐플러스 회차 제목 가져오기

```python
url3 = 'http://tv.jtbc.joins.com/replay/pr10010750/pm10047655'
html = urlopen(url3)
soup = BeautifulSoup(html, "lxml")
```

```python
print(soup.find_all('span', 'tit'))
```

```python
title_elements=soup.select('span.tit strong')
title_elements
titles=[]
for e in title_elements:
    titles.append(e.text)
titles
```

```python
import pandas as pd

name = pd.DataFrame(
    {'name': titles},
    columns={'name'})
name
```
