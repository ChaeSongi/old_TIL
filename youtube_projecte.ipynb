```python
html = []
for item in items:
    html.append('<li>')
    html.append('<a href="https://www.youtube.com/watch?v={}">'.format(item['id']['videoId']))
    html.append('<img src="{}">'.format(item['snippet']['thumbnails']['medium']['url']))
    html.append('</a>')
    html.append('<h1>{}</h1>'.format(item['snippet']['title']))
    html.append('</li>')

html
```

```python
temp = " \n ".join(html)
content = "<ul>{}</ul>".format(temp)
content[:1000]
```

```python

import webbrowser
import os
import codecs

file_path = 'youtube.html'

f = codecs.open(file_path,'w', "utf-8")

template = """
<html>
<head>
    <meta charset="UTF-8">
    <title>{{ title }}</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>
    <header>
        <h1>{{ title }}</h1>
    </header>
    <section>
        {{ content }}
    </section>
    <footer>ⓒ 장난감프로젝트</footer> 
</body>
</html>
"""

template = template.replace('{{ title }}', '송이의 API 수집 영상')
template = template.replace('{{ content }}', content)


f.write(template)
f.close()
```
