```python
import folium
from folium.features import CustomIcon
geo_df = test

map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

url = 'http://leafletjs.com/examples/custom-icons/{}'.format
icon_image = url('leaf-red.png')
shadow_image = url('leaf-shadow.png')

icon = CustomIcon(
    icon_image,
    icon_size=(38, 95),
    icon_anchor=(22, 94),
    shadow_image=shadow_image,
    shadow_size=(50, 64),
    shadow_anchor=(4, 62),
    popup_anchor=(-3, -76)
)

for i, row in geo_df.iterrows():
    mall_name = row['Name'] + '-' + row['Address']
    if row['Category'] == '롯데몰':
        icon = folium.Icon(color='red',icon='info-sign')
    elif row['Category'] == '스타필드':
        icon = folium.Icon(color='blue',icon='info-sign')
    else:
        icon = folium.Icon(color='green',icon='info-sign')

    folium.Marker(
        [row['Latitude'], row['Longitude']], 
        popup=mall_name,
        icon=icon
    ).add_to(map)

    
map.choropleth(geo_data=geo_str,
          data=guDat,
          columns=['Gu','counts'],
          fill_color='PuRd', #PuRd, YlGnBu
          key_on='feature.properties.name',
          highlight=True
          )
map
```

# 간소화하기

```python
import folium
geo_df = test

map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')

map.choropleth(geo_data=geo_str,
              data=guDat,
              columns=['Gu','counts'],
              fill_color='GnBu', #PuRd, YlGnBu
              key_on='feature.properties.name',
              highlight=True
              )

for i, row in geo_df.iterrows():
    mall_name = row['Name'] + '-' + row['Address']
    if row['Category'] == '롯데몰':
        icon_color='red'      
    elif row['Category'] == '스타필드':
        icon_color='blue'
    else:
        icon_color='green'
        
    
    icon = folium.Icon(color=icon_color,icon='info-sign')
        
    if row['Category'] == '롯데백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=700,
            color='black',
            weight=1,
            fill_opacity=0.8,
            opacity=1,
            fill_color='firebrick', # 저명도 red
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '신세계백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=700,
            color='black',
            weight=1,
            fill_opacity=0.8,
            opacity=1,
            fill_color='chocolate', # 초콜렛색
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '이마트':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=700, # 지름이 degree에 비례하도록 설정
            color='orange',
            weight=1,
            fill_opacity=0.8,
            opacity=1,
            fill_color='orange',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '현대백화점':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=700, # 지름이 degree에 비례하도록 설정
            color='orange',
            weight=1,
            fill_opacity=0.8,
            opacity=1,
            fill_color='lightgrey',
            popup=mall_name,
            fill=True
        ).add_to(map)
    elif row['Category'] == '롯데마트':
        location = (row['Latitude'], row['Longitude']) 
        folium.Circle(
            location=location,
            radius=700, # 지름이 degree에 비례하도록 설정
            color='black',
            weight=1,
            fill_opacity=0.8,
            opacity=1,
            fill_color='palevioletred',
            popup=mall_name,
            fill=True
        ).add_to(map)
    else:
        folium.Marker(
            [row['Latitude'], row['Longitude']], 
            popup=mall_name,
            icon=icon
        ).add_to(map)
    
map
```
